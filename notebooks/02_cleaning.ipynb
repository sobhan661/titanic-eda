Cleaning The Dataset

Using the information we got from looking at the dataset

1. Drop the deck column

```python
import sys
sys.path.append("..")
from data.raw.titanic import df
```

```python
# This line removes the column deck from the dataset!
df = df.drop(columns=["deck"])

# Checking to see that deck column is removed
df
```

2. Fill age missing values

```python
# This line fills the missing values in column age with the median of age
df["age"] = df["age"].fillna(value=df["age"].median())

df.isna().sum()
```

3. Fill embarked and embark_town missing values

```python
# These two lines fill the missing values in columns embarked and embark_town with thier columns first mode
df["embarked"] = df["embarked"].fillna(value=df["embarked"].mode()[0])
df["embark_town"] = df["embark_town"].fillna(value=df["embark_town"].mode()[0])
```

Checking if dataset has any missing values now

```python
df.isna().sum()
```
